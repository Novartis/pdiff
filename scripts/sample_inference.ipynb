```python
from pdiff.metadata import pDiffMetadata
from pdiff.model import pDiffModel
from pathlib import Path
```

#### 1. Load the model previously trained with sample_train_script:

```python
model_path = Path('../models/sample_training/')
my_model = pDiffModel(model_path)
```

#### 2. Get a dict of profiles to run inference on (here, the training compound profiles):

```python
test_metadata = pDiffMetadata(Path('../tests/test_data/prepared_metadata.pkl'))
test_inference_treatments = test_metadata.get_treatment_dict()
```

```python
test_inference_treatments.keys()
```

#### 3. Set appropriate GPU and transfer model

```python
!echo $CUDA_VISIBLE_DEVICES
```

```python
import torch
print(torch.cuda.is_available())
print(torch.version.cuda)
```

```python
my_model.to_device('cuda:0')
```

#### 4. Run inference and save results

```python
output_root_path = Path('../inference_test/')
results = my_model.predict(output_root_path=output_root_path, 
                 new_metadata_filename='morgan_mininvps_guidance0_inference.pkl',
                 treatment_profile_dict=test_inference_treatments,
                 gen_images_per_treatment=10,
                 random_seed=42)
```
